1. WE IMPORT THE DATA

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import euclidean_distances
plt.style.use("ggplot")
from matplotlib.pyplot import figure
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df = pd.read_csv(r"C:\Users\Administrator\Downloads\QVI_data.csv")
```

2. WE SELECT THE TRIAL STORES AND THE CONTROL STORES.
3. WE CALCULATE THE TOTAL SALES AND PRODUCT QUANTITY FROM EACH STORE.
4. WE ALSO EVALUATE THE DISTRIBUTION OF LIFESTAGE AND PREMIUM CUSTOMER FROM EACH STORE.
5. FINALLY WE JOIN INTO A SINGLE PROFILE.

```python
trial_stores = [77, 86, 88]
df_trial = df[df['STORE_NBR'].isin(trial_stores)]
df_control_candidates = df[~df['STORE_NBR'].isin(trial_stores)]
avg_sales_qty = df.groupby('STORE_NBR')[['TOT_SALES', 'PROD_QTY']].mean().reset_index()
avg_sales_qty.columns = ['STORE_NBR', 'AVG_TOT_SALES', 'AVG_PROD_QTY']


# Count number of customers per LIFESTAGE in each store
lifestage_distribution = df.groupby(['STORE_NBR', 'LIFESTAGE']).size().reset_index(name='COUNT')
lifestage_distribution['PCT'] = lifestage_distribution.groupby('STORE_NBR')['COUNT'].transform(lambda x: x / x.sum())
store_avg = df.groupby('STORE_NBR')[['TOT_SALES', 'PROD_QTY']].mean()

# Lifestage and Premium one-hot encoded proportions
lifestage_pct = df.groupby(['STORE_NBR', 'LIFESTAGE']).size().unstack().fillna(0)

lifestage_pct = lifestage_pct.div(lifestage_pct.sum(axis=1), axis=0)

# Premium one-hot encoded proportions
premium_pct = df.groupby(['STORE_NBR', 'PREMIUM_CUSTOMER']).size().unstack().fillna(0)
premium_pct = premium_pct.div(premium_pct.sum(axis=1), axis=0)

# Combine all into a single profile
store_profiles = store_avg.join(lifestage_pct).join(premium_pct)

# Fill missing columns in case some stores lack certain categories
store_profiles = store_profiles.fillna(0)
```

HERE WE STANDARDIZE THE FEATURES

```python
# Standardize the features
scaler = StandardScaler()
store_profiles_scaled = pd.DataFrame(
    scaler.fit_transform(store_profiles),
    index=store_profiles.index,
    columns=store_profiles.columns
)

trial_stores = [77, 86, 88]
control_candidates = store_profiles_scaled.drop(index=trial_stores)
```

Here we find the best control store for each trial store.

```python
from sklearn.metrics.pairwise import euclidean_distances

for trial in trial_stores:
    trial_vector = store_profiles_scaled.loc[[trial]]  # Keep as DataFrame for distance calc
    distances = euclidean_distances(trial_vector, control_candidates)
    closest_idx = distances.argmin()
    best_match_store = control_candidates.index[closest_idx]

    print(f"Best control store for Trial Store {trial}: Store {best_match_store}")
```

WE THEN CONVERT THE DATE COLUMN TO THE APPROPRIATE DATETIME FORMAT

```python
df['DATE'] = pd.to_datetime(df['DATE'])# Converts DATE column to datetime format
df['WEEK'] = df['DATE'].dt.to_period('W')
df['WEEK_START'] = df['WEEK'].apply(lambda x: x.start_time)
```

WE THEN PLOT THE TRIAL STORE AND THE RESPECTIVE CONTROL STORE TO COPMARE THEIR PERFOMANCE.
WE USE A LINE PLOT FOR THIS COMPARISON.

```python

matched_controls = {
    77: 6,
    86: 207,
    88: 237
}
for trial_store, control_store in matched_controls.items():
    # Filter for only trial and control store data
    store_data = df[df['STORE_NBR'].isin([trial_store, control_store])]
    weekly_sales = store_data.groupby(['WEEK', 'STORE_NBR'])['TOT_SALES'].sum().reset_index()

    # Convert 'WEEK' from Period to datetime (start of the week)
    weekly_sales['WEEK'] = weekly_sales['WEEK'].apply(lambda x: x.start_time)

    # Now you can safely plot
    sns.lineplot(data=weekly_sales, x='WEEK', y='TOT_SALES', hue='STORE_NBR')
    plt.xticks(rotation=45)
    plt.title(f'Trial vs Control Weekly Sales')
    plt.tight_layout()
    plt.show()
```

FROM THE PLOTS, THE MARKETING CAMPAIGNS SEEM TO BE INCREASING THE SALES FOR STORE 77 AND 86. STORE 88 ALSO HAS INCREASED SALES BUT THE IMPROVEMENT IS NOT CONSISTENTLY STRONG.
